```python
import pandas as pd
import requests
import time

# Load the existing CSV file into a DataFrame
existing_df = pd.read_csv('movie_data.csv')

# Drop unrelated columns if they exist
if 'backdrop_path' in existing_df.columns:
    existing_df = existing_df.drop(columns=['backdrop_path', 'poster_path', 'video', 'overview'])

# Define the batch size
batch_size = 100

# Initialize a variable to track the number of rows processed
processed_rows = 0

# Iterate over the DataFrame in batches of size 'batch_size'
for start_index in range(200, len(existing_df), batch_size):
    # Get the end index for the current batch
    end_index = min(start_index + batch_size, len(existing_df))
    
    # Process the current batch
    for index, row in existing_df.iloc[start_index:end_index].iterrows():
        # Check if the 'revenue' column for the current row is empty
        if pd.isnull(row['revenue']):
            # Construct the API request URL using the movie ID
            movie_id = row['id']
            url = f"https://api.themoviedb.org/3/movie/{movie_id}?language=en-US"
            
            headers = {
                "accept": "application/json",
                "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiJkODFmNjJiOWY4NmQ4NzA1MDgwNjJkMGNiYzYwYmIzYiIsInN1YiI6IjY1ZmE2ZDFkZmU2MzE4MDE3ZGZiYzljZCIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.kBnoPwr8asjuV9JvheX4lc8a3-btRS-Bs9lJoPOi_gE"
            }

            # Send a GET request to the API
            response = requests.get(url, headers=headers)
            
            # Check if the request was successful
            if response.status_code == 200:
                # Parse the JSON response
                movie_data = response.json()
                
                # Update the row with the new data
                existing_df.at[index, 'revenue'] = movie_data.get('revenue')
                # Check if the 'production_companies' column exists and has compatible dtype
                if 'production_companies' in existing_df.columns:
                    # Convert the data type of the column to object if necessary
                    if existing_df['production_companies'].dtype != 'object':
                        existing_df['production_companies'] = existing_df['production_companies'].astype(object)
                
                    # Update the row with the new data
                    existing_df.at[index, 'production_companies'] = [company['name'] for company in movie_data.get('production_companies', [])]
                existing_df.at[index, 'budget'] = movie_data.get('budget')
                time.sleep(0.1)
                
                processed_rows += 1
                print(f"Processed rows: {processed_rows}")
            else:
                # Print an error message if the request fails
                print(f"Failed to retrieve data for movie ID: {movie_id}")
        
        # Increment the count of processed rows

    # Save the updated DataFrame back to the CSV file after processing each batch
    existing_df.to_csv('movie_data.csv', index=False)

# Print the count of rows processed
print(f"Total rows processed: {processed_rows}")

# Print the count of rows with non-null 'revenue' values
print(f"Rows with 'revenue' values: {existing_df['revenue'].count()}")

existing_df.count()
```
